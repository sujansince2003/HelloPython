```python
# temperature conversion
def convertTemp(temp,unit):
    if unit=="C":
        return temp*(9/5)+32
    elif unit=="F":
        return (temp-32)*(5/9)
    else:
        return "invalid unit"

print(convertTemp(100,"C"))
print(convertTemp(100,"F"))
print(convertTemp(100,"K"))


# password strength checker

def passwordStrength(password):
    if(len(password)<8):
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
         return False
    if not any(char.isupper() for char in password):
         return False
    if not any(char in '!@#$%^&*' for char in password):
        return False
    return True

print(passwordStrength("sujan"))        # False — too short, no upper, no special
print(passwordStrength("sujan2"))       # False — no upper, no special
print(passwordStrength("sujanR2@#"))    # True  
print(passwordStrength("sujan@23#"))    # False — no uppercase



# total cost of items in a shoppping cart

def totalCost(items):
    total=0
    for item in items:
        total= total + item["price"]*item["quantity"]
    print(total)        


items=[
    {"name":"mac1","price":320,"quantity":2},
    {"name":"mac2","price":354,"quantity":3},
    {"name":"mac3","price":620,"quantity":4},
    {"name":"mac4","price":420,"quantity":6}
    ]
totalCost(items)


# check if string is palindrome
def is_palindrome(s):
    s=s.lower().replace(" ","")
    return s==s[::-1]

print(is_palindrome("sos"))


# factorial of a number

def cal_factorial(n):
    if(n==0):
        return 1
    else:
        return n* cal_factorial(n-1)    


print(cal_factorial(5))


```
