```python
def evenorodd(number):
    if(type(number) != int):
        return "not a number"
    if(number%2 == 0 ):
        return "even";
    else:
        return "odd"  

print(evenorodd("fds"))          
print(evenorodd(4))          
print(evenorodd(3))          
```
