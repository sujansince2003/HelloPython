pydantic is used to:

- Validate data

- Parse data into correct types

- Define clean data models (DTOs)

- Avoid manual validation

```python
# basic implementation
from pydantic import BaseModel
from typing import Optional
class Person(BaseModel):
    id:int
    name:str
    isStudent:Optional[bool]=None  #optional value . value will be None if nothing passed
    salary:int=150  #default value
sujan= Person(id="1",name="sujan")
print(sujan)   


```

```python
# Enum support

from pydantic import BaseModel
from enum import Enum

class Role(str,Enum):
    admin="ADMIN"
    user="USER"
    guest="GUEST"

class Person(BaseModel):   
    name:str
    role:Role

sujan = Person(name="sujan",admin="ADMIN")
print(sujan)

```
