```python
class Person:
    name='sujan'
    def sayName(self):
        return self.name
# to access var of class use self
p1= Person()
print(p1.name)
print(p1.sayName())
```

```python
class Person:
    name='sujan'
    def sayName(self,fname):
        self.fname=fname
        return self.fname
# to access var of class use self but if want to pass value when calling then use another param 
p1= Person()
print(p1.name)
print(p1.sayName("sujannnn"))
```
