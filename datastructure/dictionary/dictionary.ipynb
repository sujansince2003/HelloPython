```python
#merge 2 dict

dict1={"a":1,"b":2}
dict2={"b":3,"c":"5"}
dict3={**dict1,**dict2}
print(dict3)

```

```python
# iterating

student={
    "name":"sujan",
    "age":20,
    "gender":"male"
}


for value in student.keys():
    print(value)

for value in student.values():
    print(value)

print("index and value")
for index,value in student.items():
    print(index,value)
    print(student[index])
```

```python
# shallow copy
# this means if i change in one dict then the copied one should not copy that changed value

courses1={"name":"csit","uni":"iost"}
courses2= courses1 #this is not shallow copy i.e changing courses1 will change courses2 too
courses3=courses1.copy();

courses1["name"]="BCA"
print(courses2) #not shallow so course value will be bca
print(courses3)  #shallow so course value remain same what it was whem copying
```
