```python
# iteration in lists
numlist= [1,2,3,4,5,6]

for number in numlist:
    print(number)
print("loop completed")    


# with index

for index,number in enumerate(numlist):
    print(index,number)









```

```python
# lists in python
fruits=["apple","banana","mango"]
print(fruits)


# access elements
print(fruits[0])
print(fruits[-1])


# modify elements
fruits[0]="guava"
print(fruits)


# add elements
fruits.append("kiwi")
fruits.insert(1,"dragon")

print(fruits)


# remove elements

fruits.remove("banana")
fruits.pop()
print(fruits)


# length of lists

print(len(fruits))


# slice
print(fruits[0:2])
print(fruits[0:])

# getting index

print(fruits.index("dragon"))

# sort 

num=[1,4,2,3,5,2]
sortedNum=num.sort()
print(num)



```
