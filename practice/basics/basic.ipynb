```python
# Number guessing game (1 to 10)
import random
def NumberGuessGame(num):
  
    randomNumber= random.randint(0,10);
    
    if(num == randomNumber):
        return "The guess is right"
    else:
        return "Guess is wrong"


userInput = input("Enter your assumed number")
isNumber = userInput.isdigit();
if(isNumber):
    result= NumberGuessGame(int(userInput))
else:
    result= "Enter a valid number"    

print(result)






```

```python
# Calculator

def Calculate(op,num1,num2):
    allowed= {"+","-","*","/"}
    if(op not in allowed):
        return "Invalid operator"
        
    match(op):
        case "+":
            return num1+num2
        case "-":
            return num1-num2
        case "*":
            return num1*num2
        case "/":
            return num1/num2
        case _:
            return "invalid operator"

operator = input("Enter the operator (+,-,*,/): ")

    

number1 = float(input("Enter first number: "))  
number2 = float(input("Enter second number: ")) 
print(Calculate(operator,number1,number2))



```

```python
# password strenth checker

def main():
      strenghScore=0
      password= input("Enter the password")
      assume_special_chars = set("!@#$%^&*()")
      
      if len(password) >= 8:
        strenghScore += 1
      if any(ch.islower() for ch in password):
        strenghScore += 1
      if any(ch.isupper() for ch in password):
        strenghScore += 1
      if any(ch.isdigit() for ch in password):
        strenghScore += 1
      if any(ch in assume_special_chars for ch in password):
        strenghScore += 1
      if strenghScore > 3:
        print("password is strong", strenghScore)
      else:
        print("not strong", strenghScore)
            

main()








```
